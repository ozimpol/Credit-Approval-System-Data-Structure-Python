```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# Dataseti yükle
data = pd.read_csv('hmeq.csv')

# Eksik veri içeren satırları yoksay
data = data.dropna()

# One-hot encoding işlemi ile string ifadeleri sayısal verilere çevir
data = pd.get_dummies(data, columns=['REASON', 'JOB'])

# BAD verisindeki farklı değerleri say
bad_counts = data['BAD'].value_counts()

# Pasta grafiği tasarımı
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(bad_counts, labels=bad_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue'])
plt.title('Amount of BAD Values After Eliminate Empty Values')

# Miktarları pastanın üzerine ekleme
for i, (text, autotext) in enumerate(zip(texts, autotexts)):
    text.set_text(f"{bad_counts.index[i]}: {bad_counts.iloc[i]} Count")

plt.show()
```

```python
# Tahminde kullanılacak değişkenler ile hedef değişkeni birbirinden ayır
X = data.drop('BAD', axis=1)
y = data['BAD']

# Datayı train ve test olmak üzere 2ye ayır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# En iyi parametre hesabı için seçenekleri belirle
param_grid = {
    'hidden_layer_sizes': [(100,), (50, 50), (50, 100, 50)],  # İlgili katmanlardaki gizli nöron sayıları
    'activation': ['logistic', 'relu'],  # Aktivasyon fonksiyonları
    'solver': ['adam', 'sgd'],  # Optimizasyon algoritmaları
    'alpha': [0.0001, 0.05],  # L2 regularization parametresi
    'learning_rate': ['constant', 'adaptive'],  # Öğrenme oranı stratejileri
}

# neural network modeli oluştur
mlp_model_x = MLPClassifier(max_iter=1000)

# GridSearchCV
grid_search = GridSearchCV(mlp_model_x, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

# dataya grid işlemi uygula
grid_search.fit(X_train, y_train)

# en iyi parametreleri çıktı olarak ver
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
# farklı learning rateleri belirle
learning_rates = [0.001, 0.003, 0.01, 0.03, 0.1]
train_acc = []
test_acc = []

# her bir learning rate için
for rate in learning_rates:
    # en iyi parametreler ile MLPClassifier modeli oluştur
    mlp_model = MLPClassifier(**best_params, max_iter=1000)
    
    # modeli eğit
    mlp_model.fit(X_train, y_train)
    
    # tahmin yap
    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    
    # doğruluğu hesapla
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    train_acc.append(train_accuracy)
    test_acc.append(test_accuracy)

    # Grafik ile görselleştir
    plt.plot(mlp_model.loss_curve_, label=f'Learning Rate: {rate}')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Learning Curves for Different Learning Rates')
plt.show()
```

```python
# sayısal verileri daha kolay okuyabilmek için çıktı olarak al
results = pd.DataFrame({'Learning Rate': learning_rates, 'Training Accuracy': train_acc, 'Testing Accuracy': test_acc})
print("Results without feature scaling:")
print(results)
```

```python
# scaling işlemini uygula
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

train_acc_scaled = []
test_acc_scaled = []

for rate in learning_rates:
    # MLPClassifier modeli oluştur
    mlp_model_scaled = MLPClassifier(learning_rate_init=rate, **best_params, max_iter=1000)
    
    # Modeli eğit
    mlp_model_scaled.fit(X_train_scaled, y_train)
    
    # Tahmin yap
    y_train_pred_scaled = mlp_model_scaled.predict(X_train_scaled)
    y_test_pred_scaled = mlp_model_scaled.predict(X_test_scaled)
    
    # Doğruluk hesapla
    train_accuracy_scaled = accuracy_score(y_train, y_train_pred_scaled)
    test_accuracy_scaled = accuracy_score(y_test, y_test_pred_scaled)
    
    train_acc_scaled.append(train_accuracy_scaled)
    test_acc_scaled.append(test_accuracy_scaled)

    # Verileri görselleştir
    plt.plot(mlp_model_scaled.loss_curve_, label=f'Learning Rate: {rate}')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Learning Curves for Different Learning Rates (Scaled Features)')
plt.show()
```

```python
# Verileri daha kolay okuyabilmek için çıktı olarak al
results_scaled = pd.DataFrame({'Learning Rate': learning_rates, 'Training Accuracy': train_acc_scaled, 'Testing Accuracy': test_acc_scaled})
print("\nResults with feature scaling:")
print(results_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
import numpy as np

data = pd.read_csv('hmeq.csv')

data = data.dropna()

data = pd.get_dummies(data, columns=['REASON', 'JOB'])

X = data.drop('BAD', axis=1)
y = data['BAD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100], 
    'penalty': ['l1', 'l2'],               
    'solver': ['liblinear', 'saga']       
}

# Logistic Regression modeli oluştur
logistic_model_x = LogisticRegression(max_iter=1000)

grid_search = GridSearchCV(logistic_model_x, param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python

learning_rates = [0.001, 0.003, 0.01, 0.03, 0.1]
train_acc = []
test_acc = []

for rate in learning_rates:
    # en iyi parametrelerle logistic regression modeli oluştur
    logreg_model = LogisticRegression(**best_params, max_iter=1000)

    logreg_model.fit(X_train, y_train)

    y_train_pred = logreg_model.predict(X_train)
    y_test_pred = logreg_model.predict(X_test)

    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    
    train_acc.append(train_accuracy)
    test_acc.append(test_accuracy)

    train_sizes, train_scores, test_scores = learning_curve(logreg_model, X_train, y_train, cv=5, train_sizes=np.linspace(0.1, 1.0, 10))
    plt.plot(train_sizes, np.mean(train_scores, axis=1), label=f'Learning Rate: {rate}')

plt.xlabel('Training Examples')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Learning Curves for Different Learning Rates (Without Scaling)')
plt.show()
```

```python
results = pd.DataFrame({'Learning Rate': learning_rates, 'Training Accuracy': train_acc, 'Testing Accuracy': test_acc})
print("Results without feature scaling:")
print(results)
```

```python
# Prepare a table for training and testing accuracies for different learning rates (scaled features)
results_scaled = pd.DataFrame({'Learning Rate': learning_rates, 'Training Accuracy': train_acc_scaled, 'Testing Accuracy': test_acc_scaled})
print("\nResults with feature scaling:")
print(results_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
import joblib
import pickle
#Çıktıda uyarıların gözükmemesi için
import warnings
warnings.filterwarnings("ignore")

# Veri setini yükle
data = pd.read_csv('hmeq.csv')

# Eksik verileri içeren satırları kaldır
data.dropna(inplace=True)

# Kategorik değişkenleri sayısal değerlere dönüştürmek için one-hot encoding kullan
data = pd.get_dummies(data, columns=['REASON', 'JOB'])

# Sayısal değişkenleri standartlaştırmak için bir scaler oluştur
scaler = StandardScaler()

# Belirli bir ölçekleme faktörü (0.03) ile sayısal değişkenleri standartlaştır
scaled_values = scaler.fit_transform(data[['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC',
                                           'REASON_DebtCon', 'REASON_HomeImp', 'JOB_Mgr', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']])
scaled_values *= 0.001

# Standartlaştırılmış değerleri veri setine uygula
data[['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC',
      'REASON_DebtCon', 'REASON_HomeImp', 'JOB_Mgr', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']] = scaled_values

# Veriyi BAD=0 ve BAD=1 olmak üzere iki gruba ayır
data_bad_0 = data[data['BAD'] == 0]
data_bad_1 = data[data['BAD'] == 1]

# Her iki sınıftan eşit sayıda örnek örnekle
min_class_samples = min(len(data_bad_0), len(data_bad_1))
data_balanced = pd.concat([data_bad_0.sample(min_class_samples, random_state=42), data_bad_1.sample(min_class_samples, random_state=42)])

# Bağımsız değişkenleri ve hedef değişkeni ayır
X = data_balanced.drop('BAD', axis=1)
y = data_balanced['BAD']

# Veriyi eğitim ve test setlerine ayır
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# SMOTE ile eğitim verisini dengele
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)

# Scaler'ı bir dosyaya kaydet
joblib.dump(scaler, 'scaler.pkl')

param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 100, 50)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'adaptive'],
    'solver': ['sgd', 'adam']
}

grid_search = GridSearchCV(MLPClassifier(max_iter=1000, random_state=42), param_grid, cv=5)
grid_search.fit(X_res, y_res)

print("Best Parameters:", grid_search.best_params_)
```

```python
# En iyi modeli al
best_mlp_model = grid_search.best_estimator_

# model eğitimini yap
best_mlp_model.fit(X_res, y_res)

# tahmin yap
y_pred_mlp = best_mlp_model.predict(X_test)

# Model istatistikleri
print('MLP Model with Best Parameters:')
print('Accuracy:', accuracy_score(y_test, y_pred_mlp))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_mlp))
print('Classification Report:\n', classification_report(y_test, y_pred_mlp))

# Eğitilmiş modeli bir dosyaya kaydet
with open('mlp_model.pkl', 'wb') as f:
    pickle.dump(best_mlp_model, f)
```

```python
import warnings
warnings.filterwarnings("ignore")

# Veri setini yükle
data = pd.read_csv('hmeq.csv')

# Eksik verileri içeren satırları kaldır
data.dropna(inplace=True)

# string içeren satırları numerize et
data = pd.get_dummies(data, columns=['REASON', 'JOB'])

# numerik verileri standardize et
scaler = StandardScaler()
data[['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC',
      'REASON_DebtCon', 'REASON_HomeImp', 'JOB_Mgr', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']] = scaler.fit_transform(data[['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC',
      'REASON_DebtCon', 'REASON_HomeImp', 'JOB_Mgr', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']])

# BAD verisi 0 veya 1 olan satırları ayır
data_bad_0 = data[data['BAD'] == 0]
data_bad_1 = data[data['BAD'] == 1]

# hem 0 hem 1 olan verilerden eşit sayıda satır seç ve modeli bu satırlarla eğitmek üzere hazırla
min_class_samples = min(len(data_bad_0), len(data_bad_1))
data_balanced = pd.concat([data_bad_0.sample(min_class_samples, random_state=42), data_bad_1.sample(min_class_samples, random_state=42)])

X = data_balanced.drop('BAD', axis=1)
y = data_balanced['BAD']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)


param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  
    'penalty': ['l1', 'l2'],              
    'solver': ['liblinear', 'saga']       
}
grid_search = GridSearchCV(LogisticRegression(random_state=42), param_grid, cv=5)
grid_search.fit(X_res, y_res)

print("Best Parameters:", grid_search.best_params_)
```

```python
best_logreg_model = grid_search.best_estimator_

# modeli eğit
best_logreg_model.fit(X_res, y_res)

#Tahminleri yap

y_pred_logreg = best_logreg_model.predict(X_test)

# Model istatistikleri
print('Logistic Regression Model with Best Parameters:')
print('Accuracy:', accuracy_score(y_test, y_pred_logreg))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_logreg))
print('Classification Report:\n', classification_report(y_test, y_pred_logreg))

# Modeli bir dosyaya kaydet
with open('logreg_model.pkl', 'wb') as f:
    pickle.dump(best_logreg_model, f)
```

```python
import warnings
warnings.filterwarnings("ignore")
import joblib
import sys
from IPython.display import display, HTML
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QLineEdit, QVBoxLayout, QPushButton, QMessageBox, QComboBox

class CreditApprovalApp(QWidget):
    def __init__(self):
        super().__init__()

        self.initUI()

        # Load the scaler
        self.scaler = joblib.load('scaler.pkl')

    def initUI(self):
        self.setWindowTitle('Credit Approval Predictor')
        self.setGeometry(150, 150, 300, 300)

        self.labels = ['Loan Amount:', 'Mortgage Due:', 'Property Value:', 'Reason:', 'Job:', 'Years at Present Job:',
                       'Number of Derogatory Reports:', 'Number of Delinquent Credit Lines:', 'Age of Oldest Trade Line (in months):', 'Number of Recent Credit Inquiries', 'Number of Credit Lines', 'Debt-to-Income Ratio']

        self.input_widgets = []

        layout = QVBoxLayout()

        for label_text in self.labels:
            label = QLabel(label_text)
            if label_text in ['Reason:', 'Job:']:
                input_widget = QComboBox(self)
                input_widget.addItems(['DebtCon', 'HomeImp']) if label_text == 'Reason:' else input_widget.addItems(['Mgr', 'Office', 'Other', 'ProfEx', 'Sales', 'Self'])
            else:
                input_widget = QLineEdit(self)
            self.input_widgets.append(input_widget)

            layout.addWidget(label)
            layout.addWidget(input_widget)

        predict_button = QPushButton('Predict Approval', self)
        predict_button.clicked.connect(self.predict_approval)
        layout.addWidget(predict_button)

        self.setLayout(layout)

    def predict_approval(self):
        # Collect user input
        user_data = {}
        for label, input_widget in zip(self.labels, self.input_widgets):
            if isinstance(input_widget, QComboBox):
                if label == 'Reason:':
                    user_data['REASON_DebtCon'] = 1 if input_widget.currentText() == 'DebtCon' else 0
                    user_data['REASON_HomeImp'] = 1 if input_widget.currentText() == 'HomeImp' else 0
                elif label == 'Job:':
                    user_data['JOB_Mgr'] = 1 if input_widget.currentText() == 'Mgr' else 0
                    user_data['JOB_Office'] = 1 if input_widget.currentText() == 'Office' else 0
                    user_data['JOB_Other'] = 1 if input_widget.currentText() == 'Other' else 0
                    user_data['JOB_ProfExe'] = 1 if input_widget.currentText() == 'ProfExe' else 0
                    user_data['JOB_Sales'] = 1 if input_widget.currentText() == 'Sales' else 0
                    user_data['JOB_Self'] = 1 if input_widget.currentText() == 'Self' else 0
            else:
                user_data[label] = float(input_widget.text())

        # Load the trained models
        mlp_model = self.load_mlp_model()
        logistic_model = self.load_logistic_model()

        # Prepare user input for prediction
        user_df = pd.DataFrame([user_data])

        # Make sure the DataFrame has the same columns as the models were trained on
        model_columns = ['LOAN', 'MORTDUE', 'VALUE', 'YOJ', 'DEROG', 'DELINQ', 'CLAGE', 'NINQ', 'CLNO', 'DEBTINC',
                        'REASON_DebtCon', 'REASON_HomeImp', 'JOB_Mgr', 'JOB_Office', 'JOB_Other', 'JOB_ProfExe', 'JOB_Sales', 'JOB_Self']
        for col in model_columns:
            if col not in user_df.columns:
                user_df[col] = 0

        user_df = user_df[model_columns]

        # Apply feature scaling using the loaded scaler
        user_df_scaled = self.scaler.transform(user_df)

        # Make predictions
        mlp_prediction = mlp_model.predict(user_df_scaled)
        logistic_prediction = logistic_model.predict(user_df_scaled)

        # Display result
        result_text = f'MLP Prediction: {"Approved" if mlp_prediction[0] == 0 else "Rejected"}\n'
        result_text += f'Logistic Regression Prediction: {"Approved" if logistic_prediction[0] == 0 else "Rejected"}'

        QMessageBox.information(self, 'Result', result_text)
        self.clear_inputs()
    
    def clear_inputs(self):
        for input_widget in self.input_widgets:
            if isinstance(input_widget, QLineEdit):
                input_widget.clear()
            elif isinstance(input_widget, QComboBox):
                input_widget.setCurrentIndex(0)

    def load_mlp_model(self):

        mlp_model = joblib.load('mlp_model.pkl')
        # You need to train and save the model beforehand
        return mlp_model

    def load_logistic_model(self):

        logistic_model = joblib.load('logreg_model.pkl')
        # You need to train and save the model beforehand
        return logistic_model


if __name__ == '__main__':
    app = QApplication(sys.argv)
    credit_app = CreditApprovalApp()
    credit_app.show()
    sys.exit(app.exec_())
```
